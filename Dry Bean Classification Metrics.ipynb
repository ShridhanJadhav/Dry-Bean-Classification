```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import math
from sklearn.metrics import accuracy_score
```

```python
fv = data.iloc[:,:16]
cv = data.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(fv,cv, test_size=0.2, random_state=10, stratify=cv)
```

```python
from sklearn.preprocessing import StandardScaler,RobustScaler
```

```python
rs = RobustScaler()
rx_train= rs.fit_transform(x_train)
rx_test= rs.transform(x_test)
```

## Finding Best K value

```python
final_acc=[]
k1 = []

for k in range(1,20,2):
    
    final_acc.append(np.mean(cross_val_score(KNeighborsClassifier(n_neighbors=k), rx_train, y_train, scoring='accuracy', cv=5,)))
    k1.append(k)
    
```

```python
x = pd.DataFrame(final_acc,k1)
```

> * Best K value found to be 11

### Brute Force

```python
knn = KNeighborsClassifier(n_neighbors=11, algorithm='brute')
model = knn.fit(rx_train,y_train)
predicted = model.predict(rx_test)
accuracy = accuracy_score(y_test,predicted)
accuracy
```

### Ball_tree

```python
knn = KNeighborsClassifier(n_neighbors=11, algorithm='ball_tree')
model = knn.fit(rx_train,y_train)
predicted = model.predict(rx_test)
accuracy1 = accuracy_score(y_test,predicted)
accuracy1
```

### Kd_tree

```python
knn = KNeighborsClassifier(n_neighbors=11, algorithm='kd_tree')
model = knn.fit(rx_train,y_train)
predicted = model.predict(rx_test)
accuracy2 = accuracy_score(y_test,predicted)
accuracy2
```

## Conclusion :

> * Best K value found to be 11.
> * Experimenting with various algorithms, namely brute, ball_tree, and kd_tree, all of them resulted in the same accuracy score. Therefore, any of these algorithms can be selected for the task.

## Classification Metrics

### finding performance for each k value for cross_validation data

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score,f1_score,log_loss
```

```python
x_trainf, x_cv, y_trainf, y_cv = train_test_split(rx_train,y_train, test_size=0.2, random_state=10, stratify=y_train)
```

```python
accuracy=[]
k1= []
confusion_m=[]
presicion_m=[]
recall_m=[]
log_loss1= []
err = []
for k in range(1,20,2):
    knn = KNeighborsClassifier(n_neighbors=k)
    model = knn.fit(x_trainf,y_trainf)
    predicted = model.predict(x_cv)
    predicted_prob= model.predict_proba(x_cv)
    accuracy.append(accuracy_score(y_cv,predicted))
    confusion_m.append(confusion_matrix(y_cv,predicted))
    presicion_m.append(precision_score(y_cv,predicted, average=None))
    recall_m.append(recall_score(y_cv,predicted, average=None))
    log_loss1.append(log_loss(y_cv,predicted_prob))
    err.append(1-accuracy_score(y_cv,predicted))
    
    
    k1.append(k)
```
